```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

#### Load original dataset

```python
df = pd.read_excel('World_development_mesurement.xlsx', sheet_name='world_development')
```

#### list out development indicators

```python
features = df.columns
#df['Country'].value_counts()
```

#### Vizualise each development indicator using histogram

```python
for feature in features:
    plt.figure(figsize=(6,4))    
    sb.histplot(df[feature], kde=True)
    title = feature + " Distribution"
    plt.title(title)
    plt.show()
```

The histogram shows column 'Ease of Business','Population 15-64','Population 0-14' and 'Population Urban' normally distributed. The imputation of mean would be sugggested if missing values exists.

#### Vizualise outliers from each developement indicator using boxplot method

```python
for feature in features:
    plt.figure(figsize=(10,1))    
    #sb.histplot(df[feature], kde=True)
    sb.boxplot(df[feature], orient='h')
    title = feature + " Distribution"
    plt.title(title)
    plt.show()
```

Outliers detected in 'CO2 Emissions','Days to Start Business','Energy Usage','Health Exp % GDP','Hours to do Tax','Infant Mortality Rate','Lending Interest','Life Expectancy Female','Life Expectancy Male','Mobile Phone Usage','Population 15-64','Population 65+' and 'Population Total' columns
